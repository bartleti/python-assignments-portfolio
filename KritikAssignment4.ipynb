```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, max_iterations=1000, tolerance=1e-6):
    # Estimate the derivative of the function using symmetric approximation
    def deriv(f, base_point):
        return (f(base_point + 10**(-10)) - f(base_point - 10**(-10))) / (2 * 10**(-10))

    x_coords = [initial_point]  # List to store x_n values
    y_coords = [f(initial_point)]  # List to store f(x_n) values
    
    current_x = initial_point
    for _ in range(max_iterations):
        # Compute the derivative (gradient)
        grad = deriv(f, current_x)
        
        # Update the x-coordinate based on the gradient descent rule
        next_x = current_x - learning_rate * grad
        
        # Check for convergence (if change in x is small enough)
        if abs(next_x - current_x) < tolerance:
            break
        
        # Update current_x for the next iteration
        current_x = next_x
        
        # Store the new x and f(x) values
        x_coords.append(current_x)
        y_coords.append(f(current_x))

    # Plotting portion. You may adjust as you please.
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000)  # Plot range for smooth function curve
    function_range = [f(i) for i in plot_range]  # Compute f(x) values for the plot
    plt.plot(plot_range, function_range)  # Plot the function f(x)
    plt.plot(x_coords, y_coords, 'ro-', label="Gradient Descent Path")  # Plot the sequence of points
    
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.title("Gradient Descent")
    plt.show()

    return round(x_coords[-1], 3), round(y_coords[-1], 3)  # Return the last x_n and f(x_n), rounded to 3 decimal places

# Example usage:
# Example function: f(x) = x^2 (simple convex function)
def f(x):
    return x**2

# Perform gradient descent starting at x = 5, with a learning rate of 0.1
final_x, final_y = gradient_descent(f, learning_rate=0.1, initial_point=5)
print(f"Final x: {final_x}, Final f(x): {final_y}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, tolerance=0.001):
    # Estimate the derivative of the function using symmetric approximation
    def deriv(f, base_point):
        return (f(base_point + 10**(-10)) - f(base_point - 10**(-10))) / (2 * 10**(-10))

    # Initialize lists to store x and f(x) values
    x_coords = [initial_point]  # List to store x_n values
    y_coords = [f(initial_point)]  # List to store f(x_n) values

    # Start gradient descent iteration
    i = 0
    while abs(deriv(f, x_coords[i])) > tolerance:  # Continue until the derivative is small enough
        i += 1
        # Calculate next x using gradient descent rule
        next_x = x_coords[i-1] - learning_rate * deriv(f, x_coords[i-1])
        # Add the new x and corresponding f(x) to the lists
        x_coords.append(next_x)
        y_coords.append(f(next_x))

    # Plot the function and the path of gradient descent
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000)  # Create a smooth range for the plot
    function_range = [f(i) for i in plot_range]  # Calculate the function values for the plot range
    plt.plot(plot_range, function_range)  # Plot the function f(x)
    plt.plot(x_coords, y_coords, 'ro-', label="Gradient Descent Path")  # Plot the gradient descent path

    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.title("Gradient Descent Path")
    plt.show()

    # Return the last x and f(x), rounded to three decimal places
    return round(x_coords[-1], 3), round(y_coords[-1], 3)

# Example usage:
def f(x):
    return x**2  # Example function: f(x) = x^2 (minimum at x=0)

final_x, final_y = gradient_descent(f, learning_rate=0.1, initial_point=5)
print(f"Final x: {final_x}, Final f(x): {final_y}")
```
