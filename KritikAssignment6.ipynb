```python
import numpy as np
import matplotlib.pyplot as plt

# Normal distribution function
def fx(mu, sig, x):
    return (1 / (np.sqrt(2 * np.pi * sig))) * np.exp(-((x - mu)**2) / (2 * sig))

# Integration function using the midpoint rule
def integration(fx, a, b):
    n = 1000
    x = np.linspace(a, b, n)
    y_middle = []
    delta_x = (b - a) / n
    for i in range(n - 1):
        xm = (x[i] + x[i + 1]) / 2
        ym = fx(171, 7.1**2, xm)  # Use the squared value for variance
        y_middle.append(ym)
    ms = sum(y_middle) * delta_x
    return ms

# Function to compute the average dose
def average_dose(mean, variance, a, b, n=100000):
    dose_function = lambda x: 2.38 * x**2  # Dose calculation function
    delta_x = (b - a) / n
    points = np.linspace(a + delta_x/2, b - delta_x/2, n)
    f = fx(mean, variance, points)  # Use the fx function for the normal density
    doses = dose_function(points)  # Calculate the doses for each point
    expected_dose = np.sum(doses * f * delta_x)  # Approximate expected dose using the trapezoidal sum
    return expected_dose

# Define values for the parameters
mean = 171
variance = 7.1**2  # Standard deviation is 7.1, so variance = (7.1)^2
a = 150  # Lower bound of the range
b = 190  # Upper bound of the range

# Calculate the expected dose
expected_dose = average_dose(mean, variance, a, b)

print(f'The estimated average dose is: {expected_dose}')
```
